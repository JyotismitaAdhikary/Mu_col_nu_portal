```python
def remove_commas_in_file(file_path):
    # Read the content of the file
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Remove commas between columns
    modified_lines = [line.replace(',', ' ') for line in lines]

    # Write the modified content back to the same file
    with open(file_path, 'w') as file:
        file.writelines(modified_lines)

# Specify the path to your file
file_path = '/Users/adhikaryj/ECN3Plot/lines/Complex scalar.txt'

# Remove commas in the specified file
remove_commas_in_file(file_path)
```
