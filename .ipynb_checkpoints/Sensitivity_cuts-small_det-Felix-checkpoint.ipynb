## Define Functions to Load Events

-  this function reads the particles in a csv file and exports a data frame

```python
def load_events(filename):
        
    #read csv file
    df = pd.read_csv(filename)
    truth_energy = round(df['Energy'].loc[0],5)
    cross_section = df['Cross_section_pb'].loc[0]
    
    # loop through events and read them, fill into list 
    ievent =  -1
    particles = []
    for index, row in df.iterrows():
        
        if pd.isnull(row['pid']): 
            ievent+=1
            iparticle=0
            continue

        pid = int(row['pid'])
        px, py, pz, e = eval(row['momentum'])
        px,py,pz,e = round(px,5),round(py,5),round(pz,5),round(e,5)
        parent_pid = eval(row['parent_pids'])[0]
    
        particles.append([ievent, iparticle, truth_energy, cross_section, pid, px, py, pz, e, parent_pid])
        iparticle+=1
        
    # convert to data frame and return 
    columns=['ievent', 'iparticle', 'truth_energy', 'cross_section', 'pid', 'px', 'py', 'pz', 'e', 'parent_pid']
    data = pd.DataFrame(particles, columns=columns)
    
    return data
```

- function that evaulates all necessary obseravbles for each event

```python
def calculate_observables(data):
    events = []
    grouped_data = data.groupby('ievent')

    for ievent, evt in grouped_data:
        # Reset or initialize variables for each event
        e_mu_minus, e_mu_plus, has_charm, e_em, e_visible, ptx, pty, ht = 0, 0, 0, 0, 0, 0, 0, 0
        
        # Iterate over particles in the event
        for _, row in evt.iterrows():
            truth_energy, cross_section, pid, px, py, pz, e, parent_pid = row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9]
            
            # Check for charm particles
            if abs(parent_pid) in {411, 421, 431, 4122}: 
                has_charm = 1
            
            # Update energies for mu- and mu+
            if pid == 13: e_mu_minus = max(e, e_mu_minus)
            elif pid == -13: e_mu_plus = max(e, e_mu_plus)
            
            # Sum EM and visible energy
            if abs(parent_pid) in {22, 11}: 
                e_em += e
            if abs(parent_pid) not in {12, 14, 16}: 
                e_visible += e
                ptx += px
                pty += py
                ht += np.sqrt(px**2 + py**2)
        
        # Calculate missing pT
        pt_mis = -np.sqrt(ptx**2 + pty**2)
        
        # Append event observables
        events.append([ievent, truth_energy, cross_section, e_mu_minus, e_mu_plus, has_charm, e_visible, pt_mis, ht])

    # Construct DataFrame from results
    columns = ['ievent', 'truth_energy', 'cross_section', 'e_mu_minus', 'e_mu_plus', 'has_charm', 'e_visible', 'pt_mis', 'ht']
    observables = pd.DataFrame(events, columns=columns)
    
    return observables
```

```python
def load_everything():
    
    all_data = None
    datasets = ['NC_nue']#, 'CC_nue', 'CC_numu', 'NC_numu']
    for dataset in datasets:
        dirname = 'angle_'+dataset
        directory = pathlib.Path(dirname)
        files = sorted(directory.rglob('*GeV.csv'))
        for file in files: 
            print(str(file))
            data = load_events(str(file))
            observables = calculate_observables(data)
            observables['set'] = dataset
            if all_data is None: all_data = observables
            else: all_data = pd.concat([all_data, observables])
    return all_data
```

```python
#####New addition

rho = 7.874    ##density of Iron In grams per cm3
mass = 1.67*10**(-24)  ##mass of proton in gram
l_det = 4.0425      ## length of detector for iron for 10 kg
mul = rho*l_det/mass
nevent=10**7    ##no. of MC events can replace by ievent
def load_everything(scaling_file='MC_-14_3TeV_with_cut_with_final.txt'):
    all_data = None
    datasets = ['NC_nue', 'CC_nue', 'CC_numu', 'NC_numu']
    scaling_data = pd.read_csv(scaling_file, delim_whitespace=True) 
    for dataset in datasets:
        dirname = 'angle_' + dataset
        directory = pathlib.Path(dirname)
        files = sorted(directory.rglob('*GeV.csv'))
        
        for file in files:
            print(str(file))
            energy = float(str(file).split('_')[-1].replace('GeV.csv', ''))
            data = load_events(str(file))
            observables = calculate_observables(data)
            observables['set'] = dataset
            scaling_row = scaling_data[scaling_data['energy'] == energy]
            cross_section = scaling_row[f"{dataset}_cs"].values[0]
            flux = scaling_row[f"{dataset}_flux"].values[0]  
            scaling_factor = ((cross_section* flux * mul)/nevent)     ##scale factor 
            observables['scaling_factor'] = scaling_factor
            if all_data is None:all_data = observables
            else:all_data = pd.concat([all_data, observables])
    
    return all_data
```
